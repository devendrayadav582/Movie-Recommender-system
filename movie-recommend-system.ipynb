```python
import ast
ast.literal_eval(movies['genres'][0])
```

```python
def convert(text):
  lst=[]
  for i in ast.literal_eval(text):
      lst.append(i['name'])

  return lst
```

```python
movies['genres']=movies['genres'].apply(lambda x: convert(x))
movies['keywords']=movies['keywords'].apply(lambda x: convert(x))
```

```python
movies
```

```python
#for cast we need to make some changes in function as we want only first 3 names of cast not all
#for crew we need only directors not others
```

```python
def convert_cast(text):
  lst=[]
  counter=0
  for i in ast.literal_eval(text):
    if counter !=3:
      lst.append(i['name'])
      counter+=1
    else:
      break
  
  return lst
```

```python
movies['cast'][0]
```

```python
convert_cast(movies['cast'][0])
```

```python
def convert_director(text):
  lst=[]
  for i in ast.literal_eval(text):
    if i['job'] == 'Director':
      lst.append(i['name'])
  
  return lst
```

```python
movies['crew'][0]
```

```python
convert_director(movies['crew'][0])
```

```python
movies['cast']=movies['cast'].apply(lambda x: convert_cast(x))
movies['crew']=movies['crew'].apply(lambda x: convert_director(x))
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors=cv.fit_transform(new_df['tags']).toarray()
```

```python
#we use toarray to convert sparse matrix into array
```

```python
cv.get_feature_names()
```

```python
#here we notice one problem is some words which have same meaning are repeated 
#we use stemming to make one word from same type of words
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
  y=[]
  for i in text.split():
   y.append(ps.stem(i))

  return " ".join(y)
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors=cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors.shape
```

```python
cv.get_feature_names()
```

```python
#nees to determine distances of each movies from every other movies
```

```python
#for this we use cosine similarity
#as we know distnace and similarity have inverse realtionship
#as similarity increases distance decreases
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)
```

```python
#it array of arrays similarity
similarity
```

```python
similarity.shape
```

```python
#this show first movie similarity with all 4806 movies
similarity[0]
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x : x[1])[1:6]
```

```python
#here i made function where i provide movie name and it recommend 5 movies
```

```python
def recommend(movie):
  index=new_df[new_df['title'] == movie].index[0]
  movies_list= sorted(list(enumerate(similarity[index])),reverse=True,key=lambda x :x[1])[1:6]
  for i in movies_list:
    print(new_df.iloc[i[0]].title)
```

```python
recommend('Thor')
```
